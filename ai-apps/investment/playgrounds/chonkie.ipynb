```python
from chonkie import Visualizer, Chunk
from chonkie.genie import OpenAIGenie
from chonkie import SlumberChunker
from chonkie import RecursiveChunker
from load_dotenv import load_dotenv
import os

load_dotenv()

viz = Visualizer()

chunker = SlumberChunker(
    genie=OpenAIGenie(model="qwen/qwen3-vl-8b-instruct", api_key=os.getenv("OPENROUTER_API_KEY"), base_url="https://openrouter.ai/api/v1"),                        # Genie interface to use
    tokenizer="gpt2",  # Default tokenizer (or use "gpt2", etc.)
    chunk_size=2048,                    # Maximum chunk size
    candidate_size=196,                 # How many tokens Genie looks at for potential splits
    min_characters_per_chunk=48,        # Minimum number of characters per chunk
    verbose=True                        # See the progress bar for the chunking process
)
```

```python
f = open("t1.md", "r")

chunks = chunker.chunk(f.read())

f.close()
```

```python
viz.print(chunks)
```
